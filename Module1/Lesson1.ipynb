## üõ†Ô∏è Common Setup Issues & Fixes

| Problem | Reason | Fix |
|--------|--------|-----|
| `'python' is not recognized` | Python not in PATH | Reinstall & tick **Add to PATH** |
| ModuleNotFoundError | Missing package | Use `pip install package_name` |
| VS Code says ‚ÄúNo Python interpreter‚Äù | Interpreter not selected | Press `Ctrl+Shift+P` ‚Üí ‚ÄúPython: Select Interpreter‚Äù |
| PyCharm run error | Project interpreter missing | Set interpreter under *Settings ‚Üí Project ‚Üí Python Interpreter* |
