## ðŸ”¹ Variable
A variable is a **named storage location** that holds data in memory.

Example:
```python
age = 20


````python

# Creating variables
name = "John"
age = 21
height = 5.8
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
````

````python
# Using type() function to check data type
print(type(name))
print(type(age))
print(type(height))
print(type(is_student))
````

````python
# Examples of different data types

a = 10          # int
b = 3.14        # float
c = "Python"    # string
d = True        # boolean

print(a, "-", type(a))
print(b, "-", type(b))
print(c, "-", type(c))
print(d, "-", type(d))
````

````python
x = 10
y = 3

print("Addition:", x + y)
print("Subtraction:", x - y)
print("Multiplication:", x * y)
print("Division:", x / y)
print("Modulus:", x % y)
print("Exponent:", x ** y)
print("Floor Division:", x // y)
````

````python
a = 15
b = 10

print("a > b:", a > b)
print("a < b:", a < b)
print("a == b:", a == b)
print("a != b:", a != b)
print("a >= b:", a >= b)
print("a <= b:", a <= b)
````

````python
p = True
q = False

print("p and q:", p and q)
print("p or q:", p or q)
print("not p:", not p)
````

````python
num = 10
print("Initial:", num)

num += 5
print("After += 5:", num)

num -= 2
print("After -= 2:", num)

num *= 3
print("After *= 3:", num)

num /= 2
print("After /= 2:", num)
````

````python
# Taking user input (works in Colab too)
name = input("Enter your name: ")
age = int(input("Enter your age: "))

print("Hello", name)
print("Next year, you will be", age + 1)
````

````python
# Mini practice: Simple calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("Sum =", num1 + num2)
print("Difference =", num1 - num2)
print("Product =", num1 * num2)
print("Quotient =", num1 / num2)
````
